# Exercise: Interactive Chart in Altair

Create a **scatterplot** with dimensions of your choosing of the movies dataset that you can brush, **paired with** a stacked histogram that filters according to the brush.

See the examples at the end of Lecture 8 (Practical Data Visualization) to get started.

This is what your stacked histogram should look like: 


![Stacked Histogram](stacked_hist.png)

```python
import altair as alt
import pandas as pd

# Load the data
movies = pd.read_csv('http://vcg.github.io/upset/data/movies/movies.csv',delimiter=';')

# Melt genre columns to long format
genre_cols = ['Children', 'Documentary', 'Horror']

movies_long = movies.melt(
    id_vars=['Name', 'AvgRating', 'Watches'],
    value_vars=genre_cols,
    var_name='Genre',
    value_name='IsGenre'
)

# Keep only rows where movie belongs to that genre
movies_long = movies_long[movies_long['IsGenre'] == 1]

# Create brush selection
brush = alt.selection_interval()

# Scatterplot with brush
scatter = alt.Chart(movies_long).mark_point().encode(
    x=alt.X('Watches:Q', title='Watches'),
    y=alt.Y('AvgRating:Q', title='Average Rating'),
    color=alt.condition(brush, 'Genre:N', alt.value('lightgray'))
).add_params(
    brush
).properties(
    width=400,
    height=300,
    title='Select points to filter histogram'
)

# Stacked histogram filtered by brush
histogram = alt.Chart(movies_long).mark_bar().encode(
    x=alt.X('AvgRating:Q', bin=alt.Bin(step=1), title='AvgRating (binned)'),
    y=alt.Y('count()', title='Count of Records'),
    color='Genre:N'
).transform_filter(
    brush
).properties(
    width=400,
    height=300
)

scatter | histogram
```
